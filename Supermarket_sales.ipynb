```python
def out_lairs(col):
    Q1 = df[col].quantile(0.25)
    Q3 =  df[col].quantile(0.75)
    IQR = Q3-Q1
    data_out = df[((df[col] < (Q1-1.5*IQR)) | (df[col] > (Q3+1.5*IQR)))]
    return data_out[col] 
```

```python
out_lairs("gross income")
```
